### 🔸 Soru 12:
Bir öğrenci listesi var. Her öğrenci `(isim, not)` tuple'ı olarak saklanıyor.
- Notu 70'in üzerinde olanları filtreleyin
- En yüksek notu alan öğrenciyi bulun
- Notların ortalamasını hesaplayın

```python
# Cevabınızı buraya yazın:
ogrenciler = [
    ("Ahmet", 85),
    ("Büşra", 62),
    ("Can", 91),
    ("Deniz", 74),
    ("Elif", 58),
    ("Furkan", 88)
]
toplam=0
max_not=0
max_isim=""
for isim,not_ in ogrenciler:
  toplam+=not_
  if not_>max_not:
    max_not=not_
    max_isim=isim

  if not_>70:
    print(f"{isim}: {not_}")

print(f"Ortalama not: {toplam/len(ogrenciler)}")
print(f"En yüksek notu alan: {max_isim}: {max_not}")
```

### 🔸 Soru 13 (Set):
İki sınıfın öğrenci listesi var. `set` kullanarak:
- Her iki sınıfta da bulunan öğrencileri bulun (kesişim)
- Toplam kaç farklı öğrenci olduğunu bulun (birleşim)
- Sadece A sınıfında olup B'de olmayan öğrencileri bulun (fark)

```python
# Cevabınızı buraya yazın:
sinif_a = {"Ali", "Ayşe", "Can", "Dilan", "Ece"}
sinif_b = {"Can", "Dilan", "Fatih", "Gökçe", "Hakan"}
a=set(sinif_a)
b=set(sinif_b)
print(f"Her iki sınıfta da bulunan öğrenciler: {a.intersection(b)}")
print(f"Toplam farklı öğrenci sayısı: {len(a.union(b))}")
print(f"Sadece A sınıfında olup B'de olmayan öğrenciler: {a.difference(b)}")
```

### 🔸 Soru 15:
Özyinelemeli (recursive) bir fonksiyon yazın: `faktoriyel(n)` - n! hesaplasın.
- 0! = 1
- n! = n × (n-1)!

5!, 10! değerlerini hesaplayın.

```python
# Cevabınızı buraya yazın:
def faktoriyel(n):
  if n==0:
    return 1
  else:
    return n*faktoriyel(n-1)

print(f"5!: {faktoriyel(5)}")
print(f"10!: {faktoriyel(10)}")
```

### 🔸 Soru 16:
`lambda` fonksiyonu kullanarak:
- Bir sayının karesini hesaplayan lambda
- İki sayının büyüğünü döndüren lambda
- `sorted()` ile bir isim listesini son harfe göre sıralayan lambda

```python
# Cevabınızı buraya yazın:
isimler = ["Zeynep", "Ali", "Mahmut", "Su", "Berk"]

karesi = lambda x: x ** 2

buyuk_olan = lambda x, y: x if x > y else y

sirali_isimler = sorted(isimler, key=lambda x: x[-1])

print(f"5'in karesi: {karesi(5)}")
print(f"15 ve 8'den büyük olan: {buyuk_olan(15, 8)}")
print(f"Son harfe göre sıralı liste: {sirali_isimler}")
```

### 🔸 Soru 18 (Bütünleştirici):
Basit bir **öğrenci not sistemi** yazın:

- `ogrenciler` adında boş bir sözlük oluşturun (anahtar: isim, değer: notlar listesi)
- `ogrenci_ekle(isim)`: Yeni öğrenci ekler
- `not_ekle(isim, not_)`: Öğrenciye not ekler. Öğrenci yoksa hata mesajı verir.
- `ortalama_hesapla(isim)`: Öğrencinin not ortalamasını döndürür
- `sinif_raporu()`: Tüm öğrencileri ve ortalamalarını yazdırır, sınıf ortalamasını gösterir

En az 3 öğrenci ve her birine 3 not girerek test edin.

```python
# Cevabınızı buraya yazın:
ogrenciler = {}

def ogrenci_ekle(isim):
    if isim not in ogrenciler:
        ogrenciler[isim] = []
        print(f"{isim} sisteme eklendi.")
    else:
        print(f"{isim} isimli öğrenci zaten sistemde kayıtlı.")

def not_ekle(isim, not_):
    if isim in ogrenciler:
        ogrenciler[isim].append(not_)
        print(f"{isim} isimli öğrenciye {not_} notu eklendi.")
    else:
        print(f"HATA: {isim} isimli öğrenci bulunamadı. Lütfen önce öğrenciyi ekleyin.")

def ortalama_hesapla(isim):
    if isim in ogrenciler:
        notlar = ogrenciler[isim]
        if len(notlar) > 0:
            return sum(notlar) / len(notlar)
        else:
            return 0
    return 0

def sinif_raporu():
    print("\n" + "="*25)
    print("      SINIF RAPORU      ")
    print("="*25)

    if len(ogrenciler) == 0:
        print("Sistemde hiç öğrenci yok.")
        return

    toplam_sinif_ortalamasi = 0

    for isim in ogrenciler:
        ogr_ortalama = ortalama_hesapla(isim)
        toplam_sinif_ortalamasi += ogr_ortalama
        print(f"Öğrenci: {isim.ljust(10)} | Ortalama: {ogr_ortalama:.2f}")

    genel_ortalama = toplam_sinif_ortalamasi / len(ogrenciler)
    print("-" * 25)
    print(f"Sınıf Genel Ortalaması: {genel_ortalama:.2f}")
    print("="*25)

# Test kodu:
ogrenci_ekle("Ahmet")
ogrenci_ekle("Büşra")
ogrenci_ekle("Can")
not_ekle("Ahmet", 85)
not_ekle("Ahmet", 90)
not_ekle("Ahmet", 78)
not_ekle("Büşra", 92)
not_ekle("Büşra", 88)
not_ekle("Büşra", 95)
not_ekle("Can", 70)
not_ekle("Can", 65)
not_ekle("Can", 73)
sinif_raporu()
```
